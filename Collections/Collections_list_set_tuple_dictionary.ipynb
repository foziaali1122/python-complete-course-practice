##List methods

```python
numbers = [1, 2, 3, 4, 3, 5, 6,3]
numbers.append(56)
print("append :",numbers) #add in last
numbers.insert(1, 22)
print("insert :", numbers) #insert 22 in 1 index
numbers.remove(3)
print("remove :", numbers) #remove first 3 number
removed = numbers.pop()
print("pop :",removed) #return removed num
print(numbers)#remove end numbers and return number
numbers[0] = 21
print("modify", numbers) #modify the number
numbers.extend([2,3,4,6,8])
print("Extend:", numbers)
print("count",numbers.count(3))
print("index:",numbers.index(5))
numbers.sort()
print("sort/assicanding " ,numbers)
numbers.sort(reverse = True)
print("sort/dessicanding " ,numbers)
```

```python
tuple = (22, 45, "ali", 22)
#count
a = tuple.count(22)
print("Count:",a)
#indexing tuple
i = tuple[1]
print("Indexing:", i)
print("Slicing:", tuple[0:3])
```

###Methods in set

```python
#methods in set
x = {1, 3, 4, 3, 5, 4}
print(x)
print("length:",len(x))
x.add(22)
print(x)
x.remove(4)
print(x)
remove = x.pop()
print(remove)
print(x)
x.clear()
print(x)
```

###operations on sets

```python
#operations on sets
set1 = {1, 3, 5, 7, 9}
set2 = {1, 2, 3, 4, 5, 6}
print(set1.union(set2))  #return both set values or repeadted
print(set1.intersection(set2)) #same values of both sets
print(set1.difference(set2))  #set1 different values non in set2
print(set2.difference(set1))  #set2 different values non in set1
print(set1.symmetric_difference(set2)) #both sets different values
print(set2.difference(set1)) #all unique values
```
