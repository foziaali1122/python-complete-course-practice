##Types of Function Parameters
**1. Default Arguments:**
Default arguments allow you to define a function with pre-assigned values. If the caller does not provide an argument, the default value is used.

**2. Keyword Arguments:**
Keyword arguments allow the caller to specify arguments by name, ensuring they are matched correctly with parameters:

**3. Variable Length Arguments:**
In cases where the number of parameters is unknown, Python allows us to pass a variable number of arguments. These arguments are stored as a tuple.

```python
#default arguments
def greeting(fname = "", lname= "ali"):
  print("Hello", fname, lname)
greeting()
greeting("Uzeer")
greeting("mohsin")
```

```python
#function with keyword argument
def greeting(fname, mname, lname):
  print("Hello", fname, mname, lname)
greeting(fname ="Humza", mname ="Suhail", lname ="Lashari")
```

```python
#functions with *args (tuple of arguments) - we caanot change
def greeting(*names): #* must use
  print("Hello", names[0], names[1], names[2])
greeting("Ali", "humza", "sohail")
```

##LEGB (Local Enclosing Global Built-in ) Rule:
Python follows LEGB rule to determine the varibales scope
1. Local variable: Inside thye function.
2. Enclosing variable: Between two functions (Nasted function)
3. Global variable: Outside the function
4. Built-in variable: Predifined variable using(print ,len)

```python
x = "global"           #global variable
def outer_function():
  x = "enclosing"      #enclosing variable
  def inner_function():
    x = "local"        #local variable
    print("Inside inner function: ",x) #print local
  inner_function()
  print("Inside outer function: ",x)  #print enclosing
outer_function()
print("Outside all functions: ",x) #print global
```
