##**What Are Higher-Order Functions?**
A higher-order function is a function that:

Takes another function as an argument or

Returns a function as its result.
In simpler terms, it is a function that operates on other functions.

```python
def outer():
    def inner():
        return "Hello from inner function"
    return inner

func = outer()
print(func())
```

###**1. Map Function**
***Map use to tranform/iterate the values of iotems.***

The map function applies a given function to each item of an iterable (e.g., list, tuple) and returns a list of the results.

```python
#Function to double the numbers
def double(x):
  return x * 2

#Using map() doubled the numbers in list
numbers = [1, 2, 3,4,5]
doubled = list(map(double, numbers))
print(doubled)
```

```python
#Double of all numbers
numbers = [1, 2, 3,4,5]
result = list(map(lambda x: x*2, numbers))
print(result)
```

```python
#Square of all numbers
numbers = [1, 2, 3,4,5]
result = list(map(lambda x: x*x, numbers))
print(result)
```

```python
#cube of all numbers
numbers = [1, 2, 3,4,5]
result = list(map(lambda x: x*x*x, numbers))
print(result)
```

###**2. Filter Function**
The filter function filters elements in an iterable based on a condition defined by a function. It only returns elements that satisfy the condition.

```python
def is_even(x):
  return x % 2 == 0

numbers = [1,2,3,4,5,6,7,8]
filtered = list(filter(is_even, numbers))
print(filtered)
```

```python
#print even numbers
numbers = [1,2,3,4,5,6,7,8]
filtered = list(filter(lambda x: x % 2 ==0, numbers))
print(filtered)
```

```python
numbers = [10, 12, 15, 20, 40, 31]
filtered = list(filter(lambda x: x % 5 ==0, numbers))
print(filtered)
```

```python
#print only numbers divisible by 5
numbers = list(map(int, input("Enter numbers: ").split()))

filtered = list(filter(lambda x: x % 5 == 0, numbers))
print(filtered)
```

```python
#print greater then 10 numbers
numbers = [2, 7, 9, 8,10, 12, 15, 20, 40, 31]
filtered = list(filter(lambda x: x > 10, numbers))
print(filtered)
```

```python
#print all negative numbers
numbers = [-2, 7, -9, 8,-10, 12, -15, 20, 40, 31]
filtered = list(filter(lambda x: x < 0, numbers))
print(filtered)
```

###**3. Reduce Function**
The reduce function takes a function and an iterable, and applies the function cumulatively to the items in the iterable, returning a single output.

To use reduce, it must be imported from the functools module

```python
from functools import reduce
def multiplay(x, y):
  return x * y

#using reduce() multiplay all numbers in list
numbers = [1,2,3,4,5]
multiplied_nums = reduce(multiplay, numbers)
print(multiplied_nums)
```

```python
from functools import reduce
numbers = [1,2,3,4,5]
multiplayed_nums = reduce(lambda x, y: x * y, numbers)
print(multiplayed_nums)
```

```python
#sum 0f all numbers
from functools import reduce
numbers = [1,2,3,4,5]
multiplayed_nums = reduce(lambda x, y: x + y, numbers)
print(multiplayed_nums)
```

###**4.Zip**
Zip: Use zip when you need to pair items from multiple iterables into tuples.

```python
#using zip() pair names or ages
names = ["ali", "humza", "farhan"]
ages = [20, 13, 19]

pair = list(zip(names, ages))
print(pair)
#The zip function combines the names and ages lists into pairs of tuples.
```

### **5. Enumerate Function**
The enumerate function **adds an index** to each item in an iterable. It is useful when you need to track the index of items in a loop.

```python
fruites = {'apple', 'banana', "carrrot"}

for index , fruite in enumerate(fruites):
  print(f"Index {index} : {fruite}")
```

```python
fruites = {'apple', 'banana', "carrrot"}

for index , fruite in enumerate(fruites, start = 1): #start index from 1
  print(f"Index {index} : {fruite}")
```
