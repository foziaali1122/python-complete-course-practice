##Inheritance
*Inheritance is a way of creating a new class from an existing class*

###Type of inheritance:
* Single inheritance
* Multiple inheritance

```python
#Inheritance class

class Person:
  def __init__(self, name , age, department):
    self.name = name
    self.age = age
    self.department = department

class Teacher(Person):
  def __init__(self, name , age, department, id,course):
    super(). __init__(name , age, department)
    self.id = id
    self.course = course

  def show(self):
    print("Teacher name: ", self.name)
    print("Teacher age: ", self.age)
    print("Teacher department: ", self.department)
    print("Teacher id: ", self.id)
    print("Teacher course: ", self.course)

t = Teacher("Fozia", 18, "Ai engineer", 1002, "Ai")
t.show()
```

###Multiple Inheritance
Parent1 ------- Parent2

          |
          |
       Child

```python
class Employee:
    def __init__(self, name, company):
        self.name = name
        self.company = company

    def show(self):
        print(f"Employee: {self.name}, Company: {self.company}")

class Coder:
    def __init__(self, language):
        self.language = language

    def print_language(self):
        print(f"Language: {self.language}")

class Programmer(Employee, Coder):
    def __init__(self, name, company, language):
        Employee.__init__(self, name, company)
        Coder.__init__(self, language)

    def show_all(self):
        print(f"{self.name} works at {self.company} and codes in {self.language}")

p = Programmer("Fozia", "ICT", "Python")
p.show()
p.print_language()
p.show_all()
```

##Multilevel Inheritance :
Parent -> child -> child

```python
class Employee:
  a = 1
class programmer(Employee):
  b = 2
class manager(programmer):
  c = 3

o = Employee()
print(o.a)
#print(0.b)     #Shows an error we cannot access programmer attributes in employee
o = programmer()
print(o.a, o.b) #we can access employee attributes in programmer
o = manager()
print(o.a, o.b, o.c) #we can access employee grogrammer attributes in manager
```

###Super method
*Super method used To access the method of super class in the derived class.*

```python
class Employee:
  def __init__(self):
    print("constructor of employee")
  a = 1

class programmer(Employee):
  def __init__(self):
    print("constructor of programmer")
  b = 2
class manager(programmer):
  def __init__(self):
    print("constructor of manager")
  c = 3

o = Employee()  #run employee constructor
print(o.a)
o = programmer() #run programmer  constructor
print(o.a, o.b)
o = manager()
print(o.a, o.b, o.c) #run manager constructor
```

```python
class Employee:
  def __init__(self):
    print("constructor of employee")
  a = 1

class programmer(Employee):
  def __init__(self):
    super(). __init__()
    print("constructor of programmer")
  b = 2
class manager(programmer):
  def __init__(self):
    super().__init__()
    print("constructor of manager")
  c = 3

o = Employee()  #run employee constructor
print(o.a)
o = programmer() #run programmer and empployee constructor
print(o.a, o.b)
o = manager()
print(o.a, o.b, o.c) #run manager programmer and empployee constructor
```

```python
#Inheritance class
class Person:
    def __init__(self, name,age,Class, OM):
        self.name = name
        self.age = age
        self.Class = Class
        self.OM = OM

class Student(Person):
    def __init__(self, name,age,Class, marks_list,id):
        super(). __init__(name,age,Class, sum(marks_list)) # Pass the sum of marks as OM to Person's init
        self.id = id
        self.marks_list = marks_list # This should be a list of marks
        self.percentage = 0
        self.grade = ""

    def calculate_total(self):
        # Calculate the sum of marks in the marks_list
        self.total_marks = sum(self.marks_list)
        return self.total_marks

    def calculate_percentage(self):
        # Ensure total_marks is calculated before percentage
        self.percentage = (self.total_marks / 500) * 100 # Assuming 500 is the total possible marks
        return self.percentage

    def calculate_grade(self):
        if self.percentage >= 80:
            self.grade = "A"
        elif self.percentage >= 70:
            self.grade = "B"
        elif self.percentage >= 60:
            self.grade = "C"
        elif self.percentage >= 50:
            self.grade = "D"
        else:
            self.grade = "fail"

    def show(self):
        print("student name\t:",self.name)
        print("student age\t:",self.age)
        print("Student class\t:",self.Class)
        print("student id\t:",self.id)
        print("Student marks of 5 student:",self.marks_list)
        print("Student total marks:",self.total_marks)
        print("Student percentage:",self.percentage)
        print("Student Grade\t:",self.grade)

# Define the marks list
marks = [50,60,70,80,90]

# Create objects with the marks list
student1 = Student("Fozia",20 ,"5th",marks,"f123")

# calculate total marks, percentage and grade
student1.calculate_total()
student1.calculate_percentage()
student1.calculate_grade()

# show all info
student1.show()
```

```python
class Bankaccount:
  def __init__(self, name , amount):
    self.name = name       #public scope
    self.__amount = amount #private scope

  def show_balance(self):
    print(f"Balance : {self.__amount}")

  def deposit(self, balance):
    if balance > 0:
      self.__amount += balance #add amount to balance
      print(f"{balance} is your deposit ")
    else:
      print("Invalid deposit")

  def withdraw(self, balance):
    if 0 < balance <= self.__amount:
      self.__amount -= balance
      print(f"{balance} withdraw successfull")
    else:
      print("Invalid withdrawal")
# call objects
amount1 = Bankaccount("Fozia",6700)
amount1.show_balance()

amount1.deposit(2000)
amount1.show_balance()

amount1.withdraw(700)
amount1.show_balance()
```

## **3. Polymorphism**

One method can perform different tasks depending on input.

```python
class Animal:
  def make_sound(self):
    print("Animal sound")

class Dog(Animal):
  def make_sound(self):
    print("Dog sound: Wow wow")

class Cat(Animal):
  def make_sound(self):
    print("Cat sound: Meow Meow")

class Goat(Animal):
  def make_sound(self):
    print("Goat sound: Meh meh ")

for animal in (Dog(),Cat(),Goat()):
  animal.make_sound()
```

```python
class Animal:
  def speak(self):
    return "Animals sound"
class Dog(Animal):
  def speak(self):
    return "Dog speak bark"
class Cat(Animal):
  def speak(self):
    return "cat speak meaw"

animals = [Animal(),Dog(),Cat()]

for animal in animals:
  print(animal.speak())
```

```python
class Animal:
    def info(self):
        print("This is an animal")

class Pets:
    def pet_type(self):
        print("This is a pet")

# Multiple Inheritance
class Dog(Animal, Pets):
    def speak(self):
        print("Woof Woof!")

class Cat(Animal, Pets):
    def speak(self):
        print("Meow Meow!")

dog = Dog()
cat = Cat()

dog.speak()
cat.speak()
```

##Abstract Method
Use for the hide unnecessary code and show only important part to user

```python
from abc import ABC, abstractmethod
class Animal(ABC):
  @abstractmethod
  def speak(self):
    pass           #we just pass we cannot show animal code
class Dog(Animal):
  def speak(self):
    print("Dog speak bark")
class Cat(Animal):
  def speak(self):
    print("cat speak meaw")

dog = Dog()
cat = Cat()

dog.speak()
cat.speak()
```

```python
from abc import ABC, abstractmethod
class Vehical(ABC):
  @abstractmethod
  def start_engine(self):
    pass
  @abstractmethod
  def stop_engine(self):
    pass

class Car(Vehical):
  def start_engine(self):
     print("Car engine started")
  def stop_engine(self):
     print("Car engine stoped")

class Bike(Vehical):
  def start_engine(self):
     print("Bike engine started")
  def stop_engine(self):
     print("Bike engine stoped")

car = Car()
car.start_engine()
car.stop_engine

bike = Bike()
bike.start_engine
bike.stop_engine()
```
