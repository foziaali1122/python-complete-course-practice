##***OOPs and Methods in Python***


###**What is OOP?**

*Object-Oriented Programming (OOP) is a programming style where we solve problems by creating objects.*

**It is based on real-world concepts where objects contain:**

* Data (attributes)
* Functions (methods)

###**Class**

**Definition:**

* A class is a blueprint or template used to create objects.

* It defines what data and behavior an object will have

```python
#class is blueprint
class Employee:
  pass
```

### **Object**

**Definition:**
* An object is an instance of a class.

* It represents a real-world entity and can use class properties and methods.

```python
#class
class Employee:
  pass

#object
ali = Employee()
print(ali.language, ali.salary)

humza = Employee()
print(humza.salary, humza.language)
```

### **Attributes**

Attributes are variables inside a class.

**Types:**
* **Class Attribute →** Shared by all objects
* **Instance Attribute →** Specific to one object

```python
#class
class Employee:

  #Antributes
  language = "python"  #class attributes
  salary = 12000

#object
ali = Employee()
ali.name = "Ali"       #object attributes
print(ali.name, ali.language, ali.salary)

humza = Employee()
humza.name = "Humza"
print("\nName: ",humza.name,"\nSalary: ",humza.salary, "\nLanguage: ",humza.language)
```

```python
class car:
  name = "honda"
  colour = "blue"
  model = "node12"

car1 = car()
print(car1.name, car1.colour, car1.model)
```

### **Constructor (__init__)**

* A special method that runs automatically when an object is created.

* Used to initialize object data.

```python
class car:
  def __init__(self,name,colour,model): #init constractur
    self.name = name
    self.colour = colour
    self.model = model

car1 = car("honda", "blue", "node1")
print(car1.name, car1.colour, car1.model)
print("Car name: ",car1.name,"Car colour: ",car1.colour,"Car model: ",car1.model)
```

### **Methods**

Methods are functions inside a class.

```python
class Employee:
  language = "python"
  salary = 12000

  #Method : get information of employee
  def getInfo(self):
    print(f"language is {self.language} and salary is {self.salary}")

e = Employee()
print(e.language , e.salary)
```

###**Types of Methods:**

###**1. Instance Method**
* Works with object data
* Uses self

```python
class Empolyee:
  language = "python"
  salary = 12000

  #Method : we aslo get self
  def getInfo(self):  #using self accept the object attributes otherwise rise error
    print(f"language is {self.language} and salary is {self.salary}")

  def great(self): #without self we canot use method
    print("good morning")

harry = Empolyee()
harry.name = "fozia"
print(harry.name,harry.salary, harry.language)
harry.great()
harry.getInfo()
```

###**2.Static method**
we no need to pass method or object

```python
class Empolyee:
  language = "python" #class atrribute
  salary = 12000

  #Method : we aslo get self
  def getInfo(self):  #using self accept the object attributes otherwise rise error
    print(f"language is {self.language} and salary is {self.salary}")
  @staticmethod
  def great(): #No need self and we cannot pass object
    print("good morning")

harry = Empolyee()
harry.great()
harry.getInfo()
```

```python
#class calculator find the square , cube and squareroot of numbers and print hello
#class
class calculator:

#instractur
  def __init__(self, n):
    self.n = n

#method
  def square(self):
    print(f"Supare of {self.n} is :{self.n* self.n}")

  def cube(self):
    print(f"cube of {self.n} is :{self.n* self.n * self.n}")

  def squareroot(self):
    print(f"Supare of {self.n} is :{self.n ** 1/2}")

  @staticmethod
  def great():
    print("Hello user you are using calculator.....")

#Object
a = calculator(5)
a.great()
a.square()
a.cube()
a.squareroot()
```

###**3.Class method**
* Works with class data
* Uses @classmethod and cls

```python
#This show object instance attribute
class employee:
  a = 1
  def show(self):
    print(f"Class attribute is {self.a}")
o = employee()
o. a = 67
o.show()
```

```python
#Using @classmethod show class attribute
class employee:
  a = 1

  @classmethod
  def show(cls):
    print(f"Class attribute is: {cls.a}")
o = employee()
o. a = 67
o.show()
```

## **Operator Overloading (Magic Methods)**

Python allows operators to work with objects using dunder methods (__).

**Examples:**
* __add__() → addition (+)
* __sub__() → subtraction (-)
* __mul__() → multiplication (*)
* __truediv__() → division (/)
* __floordiv__() → floor division (//)

*This is called operator overloading.*


```python
class Number:
  def __init__(self, n):
    self.n = n

a = Number(1)
b = Number(5)

print(a+b) #Return error bcz class is accept string
```

```python
class Number:
  def __init__(self, n):
    self.n = n

  def __add__(self, num):
    return self.n + num.n

a = Number(1)
b = Number(5)

print("Addition of numbers is : ",a+b) #now its add numbers
```

```python
class Number:
  def __init__(self, n):
    self.n = n

  def __sub__(self, num):
    return self.n - num.n

a = Number(9)
b = Number(5)

print("Subtractios of numbers is : ",a-b)
```

```python
class Number:
  def __init__(self, n):
    self.n = n

  def __mul__(self, num):
    return self.n * num.n

a = Number(9)
b = Number(5)

print("Subtractios of numbers is : ",a * b)
```

```python
class Number:
  def __init__(self, n):
    self.n = n

  def __truediv__(self, num):
    return self.n / num.n

a = Number(10)
b = Number(5)

print("Divison of numbers is : ",a/b)
```

```python
class Number:
  def __init__(self, n):
    self.n = n

  def __floordiv__(self, num):
    return self.n // num.n

a = Number(10)
b = Number(5)

print("Subtractios of numbers is : ",a//b)
```

```python
#take a book class check pages
class book:
  def __init__(self, title, author, pages):
    self.title = title
    self.author = author
    self.pages = pages

  def display_show(self):
    print("\nBook title: ",self.title)
    print("Author: ", self.author)
    print("Pages: ",self.pages)

  def is_long(self):
    if self.pages >= 300:
      print(self.title, "is long book")
    else:
      print(self.title, "is short book")

book1 = book("Richvspoor", "Michal", 700)
book1.display_show()
book1.is_long()
book2 = book("24roules", "lunnardo", 200)
book2.display_show()
book2.is_long()
```

```python
class Student:
  def __init__(self, name, age, Class, rollnum):
    self.name = name
    self.age = age
    self.Class = Class
    self.rollnum = rollnum

  def show(self):
    print("\nStudent properties")
    print("Name", self.name)
    print("Age", self.age)
    print("Class", self.Class)
    print("Roll number", self.rollnum)

fozia = Student("Fozia", 18,12, 1002)
fozia.show()
aliza = Student("Aliza", 19,10, 1003)
aliza.show()
```
